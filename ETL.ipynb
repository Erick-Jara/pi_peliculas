```python
import pandas as pd
import ast
import numpy as np

from datetime import datetime

dfmovies = pd.read_csv('./dataset/movies.csv', sep = ',', encoding='latin1')

dfcredits = pd.read_csv('./dataset/credits.csv', sep = ',', encoding='latin1')
```

```python
#Convertir los valores nulos de budget y revenue en 0(cero)

dfmovies['revenue'] = pd.to_numeric(dfmovies['revenue'], errors='coerce').fillna(0)
dfmovies['budget'] = pd.to_numeric(dfmovies['budget'], errors='coerce').fillna(0)
```

```python
#Validar registros nulos en la columna release_date

df_release_date_NaN = dfmovies[dfmovies['release_date'].isna()]
df_release_date_NaN.shape[0]
```

```python
#Eliminar registros cuando release_date es nulo
print(dfmovies.shape[0])
dfmovies = dfmovies.dropna(subset=['release_date'])
print(dfmovies.shape[0])
```

```python
#Agregar columna release_year

dfmovies['release_date'] = pd.to_datetime(dfmovies['release_date'], errors='coerce')
dfmovies['release_year'] = dfmovies['release_date'].dt.year
dfmovies['release_year'] = dfmovies['release_year'].fillna(0).astype(int)
```

```python
# Eliminamos las columnas que no serán utilizadas
dfmovies = dfmovies.drop(columns=['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage', 'overview','tagline','spoken_languages','production_countries'])
```

```python
#Crear campo con el retorno de inversion
dfmovies['return'] = dfmovies.apply(lambda row: row['revenue'] / row['budget'] if row['budget'] > 0 else 0, axis=1).round(2)
```

```python
#Eliminar peliculas sin nombre
print(dfmovies.shape[0])
dfmovies = dfmovies.dropna(subset=['title'])
print(dfmovies.shape[0])
```

```python
#Eliminar peliculas sin idioma
print(dfmovies.shape[0])
dfmovies = dfmovies.dropna(subset=['original_language'])
print(dfmovies.shape[0])
```

```python
#Eliminar peliculas sin status
print(dfmovies.shape[0])
dfmovies = dfmovies.dropna(subset=['status'])
print(dfmovies.shape[0])
```

```python
dfmovies
```

```python
#desanidar belongs_to_collection

dfmovies['belongs_to_collection'] = dfmovies['belongs_to_collection'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Extraer 'id' y 'name' en nuevas columnas
dfmovies['id_collection'] = dfmovies['belongs_to_collection'].apply(lambda x: x.get('id') if isinstance(x, dict) else np.nan)
dfmovies['name_collection'] = dfmovies['belongs_to_collection'].apply(lambda x: x.get('name') if isinstance(x, dict) else '')
dfmovies['id_collection'] = dfmovies['id_collection'].fillna(0).astype(int)

dfmovies = dfmovies.drop(columns=['belongs_to_collection'])
```

```python

dfmovies['release_date'] = pd.to_datetime(dfmovies['release_date'])

def obtener_dia(fecha):
    return fecha.strftime('%A')
 

def obtener_mes(fecha):
    return fecha.strftime('%B')  

dfmovies['day'] = dfmovies['release_date'].apply(obtener_dia)
dfmovies['month'] = dfmovies['release_date'].apply(obtener_mes)

def corregir_texto(texto):
    return texto.encode('latin1').decode('latin1')

# Aplicar la función a la columna 'dias'
dfmovies['day'] = dfmovies['day'].apply(corregir_texto)
dfmovies['month'] = dfmovies['month'].apply(corregir_texto)
```

```python
dfmovies['vacios'] = dfmovies['genres'].apply(len)

condition = ((dfmovies['vacios'] == 2))
# Eliminar registros que cumplen la condición
dfmovies = dfmovies[~condition]

dfmovies
```

```python
dfmovies['return'] = pd.to_numeric(dfmovies['return'], errors='coerce')

# Eliminar filas con valores NaN en la columna 'popularity'
dfmovies = dfmovies.dropna(subset=['popularity'])

# Eliminar filas donde 'id_collection' es 0
#dfmovies = dfmovies[dfmovies['id_collection'] != 0]

# Eliminar filas con valores NaN en la columna 'return'
dfmovies = dfmovies.dropna(subset=['return'])


# Calcular la media y la desviación estándar
media = dfmovies['return'].mean()
desviacion_estandar = dfmovies['return'].std()

# Definir el límite para identificar outliers
# Puedes ajustar el factor (2 en este caso) según el nivel de sensibilidad que desees
limite_superior = media + 3 * desviacion_estandar
limite_inferior = media - 3 * desviacion_estandar

#Visualizar los outliers
dfmovies = dfmovies[~(((dfmovies['return'] < limite_inferior) | (dfmovies['return'] > limite_superior)) & (dfmovies['id_collection']==0))]
```

```python
dfmovies['popularity'] = pd.to_numeric(dfmovies['runtime'], errors='coerce')

# Eliminar filas con valores NaN en la columna 'popularity'
dfmovies = dfmovies.dropna(subset=['popularity'])

# Calcular la media y la desviación estándar
media = dfmovies['popularity'].mean()
desviacion_estandar = dfmovies['popularity'].std()

# Definir el límite para identificar outliers
# Puedes ajustar el factor (2 en este caso) según el nivel de sensibilidad que desees
limite_superior = media + 3 * desviacion_estandar
limite_inferior = media - 3 * desviacion_estandar

#outliers = dfmovies[((dfmovies['popularity'] < limite_inferior) | (dfmovies['popularity'] > limite_superior))&(dfmovies['id_collection'] == 0)]
#print(limite_inferior)
#print(limite_superior)
#outliers
#Visualizar los outliers
dfmovies = dfmovies[~(((dfmovies['popularity'] < limite_inferior) | (dfmovies['popularity'] > limite_superior))& (dfmovies['id_collection'] == 0))]
```

```python
dfmovies['popularity'] = dfmovies['popularity'].apply(lambda x: f"{x:.6f}")
```

```python
dfmovies['genres'] = dfmovies['genres'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

dfmovies['genre'] = dfmovies['genres'].apply(lambda x: ", ".join([genre['name'] for genre in x]) if isinstance(x, list) else "")

dfmovies
```

```python
#Extraer produccion

dfmovies["production"] = dfmovies["production_companies"].apply(lambda x: ", ".join([comp["name"] for comp in ast.literal_eval(x)]) if isinstance(x, str) and x.strip() else None)

dfmovies
```

```python
#df_NaN = dfmovies[dfmovies['genres'].isna()]
#df_NaN
dfmovies = dfmovies.drop(columns=['vacios', 'id_collection', 'runtime', 'production_companies', 'genres', 'status'])
dfmovies
```

```python
# Extraer los nombres de los actores y unirlos con ", "
dfcredits['cast'] = dfcredits['cast'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

dfcredits['actors'] = dfcredits['cast'].apply(lambda x: ", ".join([actor['name'] for actor in x]) if isinstance(x, list) else "")

dfcredits
```

```python
# Extraer el nombre del director
dfcredits['crew'] = dfcredits['crew'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

dfcredits['director'] = dfcredits['crew'].apply(lambda x: ", ".join([direct['name'] for direct in x if direct['job'] == 'Director']) if isinstance(x, list) else "")

dfcredits
```

```python
dfcredits = dfcredits.drop(columns=['cast', 'crew'])
dfcredits
```

```python
# Asegurar que 'id' sea de tipo entero en ambos DataFrames
dfmovies['id'] = dfmovies['id'].astype(int)
dfcredits['id'] = dfcredits['id'].astype(int)

# Fusionar SOLO cuando los valores de 'id' coincidan en ambos DataFrames
dfmovies = dfmovies.merge(dfcredits[['id', 'actors', 'director']], on='id', how='inner')

dfmovies
```
